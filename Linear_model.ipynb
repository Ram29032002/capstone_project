```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset1=pd.read_csv("Device_Monitering.csv",index_col=None)
dataset1.drop(dataset1.columns[5:198],axis=1,inplace=True)
dataset1=dataset1.drop('date',axis=1)
dataset1
```

```python
from sklearn.preprocessing import LabelEncoder

serial_encoder = LabelEncoder()
model_encoder = LabelEncoder()

dataset1['serial_number'] = serial_encoder.fit_transform(dataset1['serial_number'])
dataset1['model'] = model_encoder.fit_transform(dataset1['model'])
```

```python
indep_X=dataset1.drop(columns=['failure'])
dep_Y=dataset1['failure']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(indep_X,dep_Y,test_size=1/3,random_state=0)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
import pickle

pickle.dump(regressor, open("model.sav", "wb"))
pickle.dump(serial_encoder, open("serial_encoder.sav", "wb"))
pickle.dump(model_encoder, open("model_encoder.sav", "wb"))
pickle.dump(sc, open("scaler.sav", "wb"))
```

```python
import pickle

loaded_model = pickle.load(open("model.sav", "rb"))
serial_encoder = pickle.load(open("serial_encoder.sav", "rb"))
model_encoder = pickle.load(open("model_encoder.sav", "rb"))
sc = pickle.load(open("scaler.sav", "rb"))

serial = serial_encoder.transform(["2206E608DB42"])[0]
model = model_encoder.transform(["CT250MX500SSD1"])[0]

X = [[serial, model, 2.500590e11]]
X = sc.transform(X)

result = loaded_model.predict(X)

print(result)
```
